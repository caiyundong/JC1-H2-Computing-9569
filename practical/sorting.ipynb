```python
def insertion_sort(arr):
    for idx in range(1, len(arr)):
        unsorted_val = arr[idx]
        current = idx - 1

        while current >= 0 and unsorted_val < arr[current]:
            arr[current + 1] = arr[current] # copy the large value to 1 position at right
            current = current - 1 # check next value at left

        arr[current+1] = unsorted_val

insertion_sort(l)
print(l)
```

```python
def bubble_sort(arr):
    for i in range(len(arr) - 1):
        print(i)
        bSwap = False
        for j in range(0, len(arr) - 1 - i):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                bSwap = True

        if not bSwap:
            print("No swap, already sorted")
            break

l = [1, 5, 3, 2, 4]
l = [1, 2, 5, 4, 3]
bubble_sort(l)
print(l)
```

```python
# quicksort - in-place
# need to use swap operation
# using last as pivot
def quicksort(arr, low=0, high=None):
    if high is None:
        high = len(arr) - 1

    if low < high:
        pivot_index = partition(arr, low, high)
        quicksort(arr, low, pivot_index - 1)
        quicksort(arr, pivot_index + 1, high)

def partition(arr, low, high):
    # faster pointer
    # lower pointer
    pivot = arr[high]  # choose the last element as pivot
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

# def partition(arr, low, high):
#     # first one as pivot
#     pivot = arr[low]
#     left = low + 1
#     right = high
#     done = False

#     while not done:
#         while left <= right and arr[left] <= pivot:
#             left += 1

#         while left <= right and arr[right] >= pivot:
#             right -=1

#         if left > right:
#             break
#         else:
#             arr[left], arr[right] = arr[right], arr[left]

#     arr[low], arr[right] = arr[right], arr[low]
#     return right


l = [2, 5, 4, 1, 7, 3]
quicksort(l)
print(l)
```

```python
def mergesort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = mergesort(arr[:mid])
    right = mergesort(arr[mid:])

    return merge(left, right)

def merge(left, right):
    merged = []
    i = j = 0

    # merge two sorted lists
    # stop when running out of any two lists
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1

    # append remaining elements
    if i < len(left):
        # merged.extend(left[i:])
        merged = merged + left[i:]
    if j < len(right):
        # merged.extend(right[j:])
        merged = merged + right[j:]
    return merged


l = [2, 5, 4, 1, 7, 3]
sorted_l = mergesort(l)
print(l)
print(sorted_l)
```
